```python
import pandas as pd

def convert_age_to_days(age):
    if pd.isnull(age):
        return None
    
    value, unit = age.split()
    value = int(value)
    
    if 'year' in unit:
        return value * 365
    elif 'month' in unit:
        return value * 30
    elif 'week' in unit:
        return value * 7
    elif 'day' in unit:
        return value
    else:
        return None

# Apply conversion
df['age_in_days'] = df['age_upon_intake'].apply(convert_age_to_days)
```
